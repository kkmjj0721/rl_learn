一些简单操作

```python
import torch
import numpy as np

# 1. 创建 Tensor (跟 NumPy 一模一样)
x_np = np.array([1, 2, 3])
# 我们可以在后面加上指定设备，如：
# x_torch = torch.tensor([1, 2, 3], device='cuda') or x_torch = torch.tensor([1, 2, 3], device='cuda:0') or x_torch = torch.tensor([1, 2, 3], device='cpu') 
x_torch = torch.tensor([1, 2, 3]) 
print(x_torch)

# 2. 互相转换 (以后经常用)
# Tensor -> NumPy (画图时必须转回 NumPy)
back_to_np = x_torch.numpy()
# NumPy -> Tensor (喂给神经网络时必须转成 Tensor)
back_to_torch = torch.from_numpy(x_np)

# 3. 基础运算
y = x_torch + x_torch  # 加法
z = x_torch * 2        # 数乘
print(f"运算结果: {y}")
```

自动求导

```python
import torch

x = torch.tensor(3.0,requires_grad=True)

y = x**2 + 2*x + 1
print(f"y的值：{y}")

y.backward()
print(f"y在x点的导数为：{x.grad}")
```
