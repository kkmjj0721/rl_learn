初识智能体

```python
import gymnasium as gym
import time

# 1. 创建环境 (CartPole-v1: 倒立摆)
# render_mode="human" 会弹出一个窗口让你看到动画
env = gym.make("CartPole-v1", render_mode="human")

# 2. 重置环境，获取初始状态 (Observation)
# observation: [小车位置, 小车速度, 杆子角度, 杆子角速度]
observation, info = env.reset()

# 3、打印分析
print("-" * 30)
print(f"初始 Observation 内容: {observation}")
print(f"Observation 形状: {observation.shape}")
print("-" * 30)
print(f"Action Space 类型: {env.action_space}")
print("含义: 0=向左推, 1=向右推")
print("-" * 30)

time.sleep(3) 
env.close()
```

让小车动起来

```python
import gymnasium as gym

# 1、创建环境
env = gym.make("CartPole-v1", render_mode="human")
state, info = env.reset()
total_reward = 0.0

for step in range(100):
    # 显示画面
    env.render() 

    # 随机选择动作
    action = env.action_space.sample()

    # 推动仿真，同时获取返回值判断
    new_state, reward, terminated, truncated, info = env.step(action)

    # 计算总奖励
    total_reward += reward
    print(f"Step: {step}, Action: {action}, Reward: {reward}")

    # 判断是否终止
    if terminated or truncated:
        print("游戏结束！")
        break;

print(f"总得分:{total_reward}")
env.close()
```
